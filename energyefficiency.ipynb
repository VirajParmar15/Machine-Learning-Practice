```python
import numpy as np
import pandas as pd

df = pd.DataFrame(df)

def detect_outliers_z_score(df, threshold=3):
    z_scores = np.abs((df - df.mean()) / df.std())
    return z_scores > threshold
df = pd.read_csv('/content/drive/MyDrive/DATASETS/energy_efficiency_data.csv')

outliers_mask = detect_outliers_z_score(df)

for column in df.columns:
    print(f"Outliers in column '{column}':")
    print(df[outliers_mask[column]])
    print()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop("Cooling_Load",axis=1),
                                                 df["Cooling_Load"],
                                                 test_size=0.3,
                                                 random_state=0)
```
